```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x))
```

```python
# iteration 1
b=0
m=78.35
slope=-2 * np.sum(y-m*x.ravel()-b)
slope
```

```python
n = 0.1

step_size=n*slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred=((m*x)+b)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),label="ols")
plt.plot(x,y_pred,label="gd")
plt.legend()
```

```python
# iteration 2
slope=-2 * np.sum(y-m*x.ravel()-b)
slope

```

```python
step_size=n*slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred1=((m*x)+b)
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),label="ols",color = "red")
plt.plot(x,y_pred,label=f"gd it 1")
plt.plot(x,y_pred1,label=f"gd it 2",color = "black")
plt.legend()
```
